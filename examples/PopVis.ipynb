# Exploring flowcytometry data obtained during the 2013 deepDOM cruise
The data consists of continuous flowcytometric measurements of photosynthetic pico-eukaryotes using the SeaFlow cytometer (http://armbrustlab.ocean.washington.edu/resources/seaflow/) obtained at a transect from Uruguay to the Amazon.



## Goals

Marine phytoplankton form the base of the marine foodweb and it is important to gain insight in their occurance/abundance throughout the oceans.
In this notebook we are going to explore how the abundance of 4 different groups of pico phytoplankton cells vary over a three week transect from Uruguay to Brazil.

We will explore in particular two questions:
1) Which group of picophytoplankton is most abundant
2) Is there diel variation in cell abundance

### Question 1. Which group of picophytoplankton is most abundant

The abundance of four major picophytoplankton groups, Prochlorococcus, Synechochoccus, pico eukaryotes and 'others' is plotted against time. 
The y-ax shows the abundance expressed in 10^6 cells/L. The scale of the axes change per population. 
The most abundant population are the prochlorococcus, followed by synechococcus.
The least abundant are the pico-eukaryotes.

Note that there is considerable variation in abundance over time.

### Question 2. Is there diel variation in cell abundance

The average population abundance (aggregated over the 3 week measuring period) is plotted against time of day. This shows the diel fluctuation in cell abundance for the four populations.

All four populations show diel variation, reflecting cell division and turnover. While the prochlorococcus population show two peaks per day (in the evening and at early morning) the other 3 population only show one peak at different times in the day.
